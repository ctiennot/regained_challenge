```python
from helpers.data_loading import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
faces, _ = faces_batch(range(1, 10001), black_white=False)
y, _ = load_labels()[0:10000]
faces /= 255.
```

```python
plt.imshow(faces[15])
```

```python
for i in range(2):
    print "score: {}".format(y[i])
    f, axarr = plt.subplots(1, 4, figsize=(10, 3))
    axarr[0].hist(faces[i, :,:,0].flatten())
    axarr[1].hist(faces[i, :,:,1].flatten())
    axarr[2].hist(faces[i, :,:,2].flatten())
    axarr[3].imshow(faces[i])
    plt.show()
```

```python
for note in range(0, 25):
    print "note", note, "nb samples", sum(y==note)
    avg_beautiful = faces[y==note].mean(axis=0)
    f, axarr = plt.subplots(1, 4, figsize=(10, 3))
    axarr[0].hist(avg_beautiful[:,:,0].flatten())
    axarr[1].hist(avg_beautiful[:,:,1].flatten())
    axarr[2].hist(avg_beautiful[:,:,2].flatten())
    axarr[3].imshow(avg_beautiful)
    plt.show()
```

```python
avg_beautiful = faces[y>20].mean(axis=0)
std_beautiful  = faces[y>20].std(axis=0)
f, axarr = plt.subplots(1, 3, figsize=(10, 3))
axarr[0].imshow(avg_beautiful-std_beautiful)
axarr[1].imshow(avg_beautiful)
axarr[2].imshow(avg_beautiful+std_beautiful)
plt.show()

avg_ugly = faces[y<10].mean(axis=0)
std_ugly = faces[y<10].std(axis=0)
f, axarr = plt.subplots(1, 3, figsize=(10, 3))
axarr[0].imshow(avg_ugly-std_ugly)
axarr[1].imshow(avg_ugly)
axarr[2].imshow(avg_ugly+std_ugly)
plt.show()
```
